### information gain quiz

Which of the following splitting criteria provides the most information gain for discriminating Mobugs from Lobugs?



```python
data = pd.read_csv("../data/ml_bugs_data.csv").sort_values(by = "Species")
data
```

```python
data.Color.unique()
```

```python
len(data[data["Species"]=="Lobug"])
```

```python
def Entropy(num_lobug, num_mobug):
    p1 = num_lobug/num_lobug+num_mobug
    p2 = num_mobug/num_lobug+num_mobug
    return -p1*np.log2(p1) - p2*np.log2(p2)
```

```python
Ent_parent = Entropy(len(data[data["Species"]=="Lobug"]), len(data[data["Species"]=="Mobug"]))
Ent_parent
```

```python
def Ent_child(data, cond):
    
    child = data[cond]
    
    return Entropy(len(child[child["Species"] == "Lobug"]), len(child[child["Species"] == "Mobug"]))
```

```python
Ent_child1 = Ent_child(data, data["Color"]=="Green")
Ent_child2 = Ent_child(data, data["Color"]!="Green")

(Ent_child1, Ent_child2)
```

```python
def Info_gain(data, Ent_parent, cond1, cond2):
    Ent_child1 = Ent_child(data, cond1)
    Ent_child2 = Ent_child(data, cond2)
    
    p1 = len(data[cond1])/len(data)
    p2 = len(data[cond2])/len(data)
    
    return Ent_parent - p1*Ent_child1 - p2*Ent_child2
```

```python
Info_gain(data, Ent_parent, data["Color"]=="Green", data["Color"]!="Green")
```

```python
Info_gain(data, Ent_parent, data["Color"]=="Blue", data["Color"]!="Blue")
```

```python
Info_gain(data, Ent_parent, data["Color"]=="Brown", data["Color"]!="Brown")
```

```python
Info_gain(data, Ent_parent, data["Length (mm)"]<20.0, data["Length (mm)"]>=20.0)
```

```python
Info_gain(data, Ent_parent, data["Length (mm)"]<17.0, data["Length (mm)"]>=17.0)
```
